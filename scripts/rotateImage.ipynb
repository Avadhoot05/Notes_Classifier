```python
#Rotate images without getting cropped
import numpy as np
import os
import cv2
```

```python
#path to the dataset directory
path = "C:\\Users\\khede\\OneDrive\\Desktop\\DESKTOP\\Notes Classifier\\NonNotes"
path
```

```python
#calculating total number of training examples
imgList = os.listdir(path)
m = len(imgList)     #number of training expamples(m)
imgList
```

```python
i=337
for imgs in imgList:
    global path
    cvt = cv2.imread(path+"\\"+imgs)
    (h,w) = cvt.shape[:2]
    (cX, cY) = (w/2, h/2)
    
    M = cv2.getRotationMatrix2D((cX,cY),-45,1.0)
    
    cos = np.abs(M[0,0])
    sin = np.abs(M[0,1])
    
    nW = int((h*sin) + (w*cos))
    nH = int((h*cos) + (w*sin))
    
    M[0,2] +=(nW/2) -cX
    M[1,2] +=(nH/2) -cY
    
    rotated = cv2.warpAffine(cvt,M,(nW,nH))
    cv2.imwrite("C:\\Users\\khede\\OneDrive\\Desktop\\DESKTOP\\Notes Classifier\\rotateNotes\\"+str(i)+".jpg",rotated)
    i+=1
```
