```python
#this folder renames files in alphbetical manner
import os
import numpy as np
```

```python
Sourcedir_path = "C://Users//khede//OneDrive//Desktop//DESKTOP//Notes Classifier//NonNotes"+"//"
Targetdir_path = "C://Users//khede//OneDrive//Desktop//DESKTOP//Notes Classifier//target2"+"//"
```

```python
img_list = os.listdir(Sourcedir_path)
img_list
```

```python
#first folder rename
s = ' '
i=0
no_of_img = len(img_list)


c=0

while True:
    alphabet ='A'
    while True:
        strn =(s+alphabet).strip()
        os.rename(Sourcedir_path+img_list[c], Targetdir_path+strn+".jpg")
        c+=1
        alphabet =chr(ord(alphabet)+1)
        i +=1
        if strn[-1] == 'Z' or i == no_of_img:
            break
    if i == no_of_img:
        break
    s=strn

```

```python
#after first folder

#IF LAST FILE IN PREVIOUS FOLDER IS ZZZZZZZZZZZZR.jpg

s = 'ZZZZZZ'    #SPECIFY NAME OF LAST FILE IN PREVIOUS FOLDER EXCLUDING LAST CHARACTER
i=0
no_of_img = len(img_list)

alphabet ='M'        #SPECIFY LAST CHARACTER + 1

c=0

while True:
    while True:
        strn =(s+alphabet).strip()
        os.rename(Sourcedir_path+img_list[c], Targetdir_path+strn+".jpg")
        c+=1
        alphabet =chr(ord(alphabet)+1)
        i +=1
        if strn[-1] == 'Z' or i == no_of_img:
            break
    if i == no_of_img:
        break
    s=strn
    alphabet ='A'
```
